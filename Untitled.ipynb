```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
first = cv2.imread('bird.jpg')
```

```python
plt.title('original')
plt.imshow(first)
```

```python
crop_img = first[0:700,400:850]
```

```python
plt.title('crop image')
plt.imshow(crop_img)
```

```python
second = cv2.imread('nature.jpg')
```

```python
plt.title('original_nature_image')
plt.imshow(second)
```

```python
crop_2=second[0:700,400:850]
```

```python
plt.title('cropped_nature_image')
plt.imshow(crop_2)
```

```python
merge=np.hstack((crop_img,crop_2))
plt.title('_nature_image')
plt.imshow(merge)
```
